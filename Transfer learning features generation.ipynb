```python
from trainer import Trainer
from resnet import fineTuneResNet50,lessFilterResNet50, resnet50
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_json('Data/denoised_processed_train.json')
print(train_df.shape)
```

```python
print(np.array(train_df.ix[0, 'band_mixed']).shape)
```

## Collect features for transfer learning

```python
model = resnet50(pretrained=True)

tr = Trainer(
        model,
        None,
        epochs=900,
        milestones=[50, 600],
        gamma=0.1,
        batch_size=128, 
        use_cuda=True, 
        gpu_idx=0)

# is_general_model=True ==> takes only one input to the model
train_pred = tr.test(train_df, is_transfer_learning=True, is_general_model=True)
```

```python
print(train_pred.shape)
```

```python
transfer_learning_train = pd.DataFrame(train_pred, columns=['feature_'+str(i) for i in range(1000)])
transfer_learning_train['is_iceberg'] = train_df['is_iceberg']
print(transfer_learning_train.head(3))
```

```python
transfer_learning_train.to_csv('Data/transfer_learning_from_ResNet5_train.csv', float_format='%.6f', encoding='utf-8', index=False)

del transfer_learning_train, train_df
```

```python
test_df = pd.read_json('Data/denoised_processed_test.json')
print(test_df.shape)
test_pred = tr.test(test_df, is_transfer_learning=True, is_general_model=True)
print(test_pred.shape)

transfer_learning_test = pd.DataFrame(test_pred, columns=['feature_'+str(i) for i in range(1000)])
transfer_learning_test['is_iceberg'] = test_df['is_iceberg']
print(transfer_learning_test.head(3))

transfer_learning_test.to_csv('Data/transfer_learning_from_ResNet5_test.csv', float_format='%.6f', encoding='utf-8', index=False)
del transfer_learning_test, test_df
```

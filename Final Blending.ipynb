```python
outs = [pd.read_csv('Submissions/'+f, index_col=0) for f in best_files]
concat_sub = pd.concat(outs, axis=1)
cols = list(map(lambda x: "is_iceberg_" + str(x), range(len(concat_sub.columns))))
concat_sub.columns = cols
concat_sub.reset_index(inplace=True)
concat_sub.head(5)
```

```python
# set up cutoff threshold for lower and upper bounds, easy to twist 
cutoff_lo = 0.67
cutoff_hi = 0.33
```

```python
import numpy as np
from scipy.stats import trim_mean
# get the data fields ready for stacking
concat_sub['is_iceberg_max'] = concat_sub.iloc[:, 1:len(files)].max(axis=1)
concat_sub['is_iceberg_min'] = concat_sub.iloc[:, 1:len(files)].min(axis=1)
concat_sub['is_iceberg_mean'] = concat_sub.iloc[:, 1:len(files)].mean(axis=1)
concat_sub['is_iceberg_median'] = concat_sub.iloc[:, 1:len(files)].median(axis=1)
concat_sub['25_percentile'] = np.percentile(concat_sub.iloc[:, 1:len(files)], 50, axis=1, keepdims=True)
concat_sub['75_percentile'] = np.percentile(concat_sub.iloc[:, 1:len(files)], 75, axis=1, keepdims=True)
concat_sub['trim_mean_01'] = trim_mean(concat_sub.iloc[:, 2:len(files)], 0.1, axis=1) # except best file
concat_sub['trim_mean_02'] = trim_mean(concat_sub.iloc[:, 2:len(files)], 0.2, axis=1) # except best file
```

```python
from scipy.stats import gmean
import numpy as np
concat_sub['is_iceberg_gmean'] = gmean(concat_sub.iloc[:, 1:len(files)], axis=1)
```
